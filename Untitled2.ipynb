```python
import os, sys
 
# set SDL to use the dummy NULL video driver, 
#   so it doesn't need a windowing system.
#os.environ["SDL_VIDEODRIVER"] = "dummy"
 
 
import pygame.transform

#some platforms might need to init the display for some parts of pygame.
import pygame.display
pygame.display.init()
screen = pygame.display.set_mode([512,512]) 
 
 
def scaleit(fin, fout, w, h):
    i = pygame.image.load(fin)
    w2 , h2 = i.get_width(),i.get_height()
    s = pygame.Surface([w2,h2])
    s.blit(i,(0,0),(0,0,w2,h2))
    
    pygame.image.save(s, fout)
 
 
if __name__ == "__main__":
    fin, fout, w, h = '/tmp/test.png','/tmp/test2.png',200,200
    w, h = map(int, [w,h])
    scaleit(fin, fout, w,h)
```
